```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnt = 0
df_dict = {}
for city in ['Albany']:
    if not os.path.isdir(f"../final/{city}") or city == "README.md":
        continue
    files = os.listdir(f"../final/{city}")
    if len(files) == 0:
        continue
    for file in files:
        try:
            df = pd.read_csv(f"../final/{city}/{file}")
            df_dict[file] = df
            break
        except Exception as e:
            print(f"Error reading {file} in {city}: {e}")
            continue
    break
    
    cnt += 1
    
```

```python
df = df_dict['calendar.csv'].copy()
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
df.groupby('listing_id')['price'].agg(['mean', 'std']).sort_values('std', ascending=False)
```
